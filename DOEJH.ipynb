# **Ingresando los datos a Python**

```python
from plotnine.geoms import geom_point
df = pd.DataFrame(data={"Sexo":["Masculino","Masculino", "Femenino", "Femenino", "Masculino","Masculino", "Femenino", "Femenino","Masculino","Masculino", "Femenino", "Femenino","Masculino", "Femenino",], 
                        "Estatura (cm)": [183, 179, 160, 172, 180, 170, 179, 160, 162, 170, 175, 165, 162, 155 ],
                        "Peso (kg)":[82, 75.1, 50, 58.7, 72, 80, 60, 67, 80, 74, 55, 59, 75, 80]})
```

# **Visualizando los datos y realizando el diagrama de puntos**

```python
print(df)
print(p9.ggplot(df)+p9.aes(x="Estatura (cm)", y="Peso (kg)", color = "Sexo")+p9.geom_point())
```

# **Realizando el Diagrama de Caja de las varianbles Peso y Estatura con sexo**

```python
(p9.ggplot(df)+p9.aes(x="Estatura (cm)", y="Peso (kg)", fill = "Sexo")+p9.geom_boxplot())
```

# **Grafiquemos la función Densidad para Estatura con el Sexo**

```python
(p9.ggplot(df)+p9.aes(x="Estatura (cm)", fill = "Sexo")+p9.geom_density(alpha = 0.5))
```

# **Grafiquemos la función Densidad para Peso con el Sexo**

```python
(p9.ggplot(df)+p9.aes(x="Peso (kg)", fill = "Sexo")+p9.geom_density(alpha = 0.5))
```
